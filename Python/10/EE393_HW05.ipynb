```python
#Batu Ozmeteler S009924 EE Department
import math
import numpy as np
import scipy.integrate as scint
import matplotlib.pyplot as plt

def midpoint(f, a, b, n):
    h = (b-a)/n
    sum = 0
    for i in range(0,n):
        x = a + i*h + h/2
        sum += f(x)
    return sum*h 

actual1 = 2
actual2 = 0
actual3 = 2
actual4 = 1

sciPy1 = scint.quad(np.exp, 0, np.log(3))[0]
sciPy2 = scint.quad(np.cos, 0, np.pi)[0]
sciPy3 = scint.quad(np.sin, 0, np.pi)[0]
sciPy4 = scint.quad(np.sin, 0, np.pi / 2)[0]

midPoint1 = midpoint(lambda x: math.exp(x), 0, math.log(3), 100)
midPoint2 = midpoint(lambda x: math.cos(x), 0, math.pi, 100)
midPoint3 = midpoint(lambda x: math.sin(x), 0, math.pi, 100)
midPoint4 = midpoint(lambda x: math.sin(x), 0, math.pi / 2, 100)

print("RESULTS OBTAINED FROM EACH IMPLEMENTATION \n")
print("The integral of exp(x) in range[0,log3] is " + str(sciPy1) + 
      " with sciPy implementation, " + str(midPoint1) + 
      " with midpoint implementation. But the actual value is " + str(actual1) + "\n")

print("The integral of cos(x) in range[0,π] is " + str(sciPy2) + 
      " with sciPy implementation, " + str(midPoint2) + 
      " with midpoint implementation. But the actual value is " + str(actual2) + "\n")

print("The integral of sin(x) in range[0,π] is " + str(sciPy3) + 
      " with sciPy implementation, " + str(midPoint3) + 
      " with midpoint implementation. But the actual value is " + str(actual3) + "\n")

print("The integral of sin(x) in range[0,π/2] is " + str(sciPy4) + 
      " with sciPy implementation, " + str(midPoint4) + 
      " with midpoint implementation. But the actual value is " + str(actual4) + "\n")

print("PERCENTAGE AND ABSOLUTE ERRORS OF EACH IMPLEMENTATION \n")
print("Percentage errors of scipy implementation are %" +  str(abs((((sciPy1/actual1)-1)*100))) + 
      ", can't be calculated since the " + "actual result is zero, %" + str(abs((((sciPy3/actual3)-1)*100))) + 
      " and %" + str(abs((((sciPy4/actual4)-1)*100))) + " respectively.\n")

print("Percentage errors of midpoint implementation are %" +  str(abs((((midPoint1/actual1)-1)*100))) + 
      ", can't be calculated since the " + "actual result is zero, %" + str(abs((((midPoint3/actual3)-1)*100))) + 
      " and %" + str(abs((((midPoint4/actual4)-1)*100))) + " respectively.\n")

print("Absolute errors of scipy implementation are " + str(actual1 - sciPy1) + ", " + str(actual2 - sciPy2) + 
      ", " + str(actual3 - sciPy3) + ", " + str(actual4 - sciPy4) + " respectively.\n")

print("Absolute errors of midpoint implementation are " + str(actual1 - midPoint1) + ", " + str(actual2 - midPoint2) + 
      ", " + str(actual3 - midPoint3) + ", " + str(actual4 - midPoint4) + " respectively.\n")

print("GRAPHS OF EACH FUNCTION IN THE INDICATED RANGES \n")
xValues =  np.linspace(0, 4, 100)

xAxis1 = []
xAxis2 = []
xAxis3 = []
xAxis4 = []

yAxis1 = []
yAxis2 = []
yAxis3 = []
yAxis4 = []

for xValue in xValues: 
    yValue1 = math.exp(xValue)
    yValue2 = math.cos(xValue)
    yValue3 = math.sin(xValue)
    yValue4 = math.sin(xValue)
    if(xValue >= 0 and xValue <= math.log(3)):
        yAxis1.append(yValue1)
        xAxis1.append(xValue)
    if(xValue >= 0 and xValue <= math.pi):
        yAxis2.append(yValue2)
        xAxis2.append(xValue)
    if(xValue >= 0 and xValue <= math.pi):
        yAxis3.append(yValue3)
        xAxis3.append(xValue)
    if(xValue >= 0 and xValue <= math.pi/2):
        yAxis4.append(yValue4)
        xAxis4.append(xValue)

plt.plot(xAxis1,yAxis1,label='exp(x)')
plt.plot(xAxis2,yAxis2,label='cos(x)')
plt.plot(xAxis3,yAxis3,label='sin(x)')
plt.plot(xAxis4,yAxis4,label='sin(x)')
plt.legend()
plt.gca().xaxis.grid(True) 
plt.gca().yaxis.grid(True)
plt.show()
```
